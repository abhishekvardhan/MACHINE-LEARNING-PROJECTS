```python
import pandas as pd
from selenium import webdriver
import os
os.chdir("C:\\Users\\KARNDEEP SINGH\\Desktop\\data scraping\\geckodriver-v0.25.0-win64")
browser = webdriver.Firefox("C:\\Users\\KARNDEEP SINGH\\Desktop\\data scraping\\geckodriver-v0.25.0-win64")
```

```python
from bs4 import BeautifulSoup
```

```python
master_data ={}
for year in range(2013,2019):
    Average_Temp = []
    Max_Temp = []
    Min_Temp = []
    Atm_Pressure = []
    Average_relative_humidity = []
    Total_rainfall_snowmelt= []
    Average_visibility= []
    Average_wind_speed= []
    Maximum_sustained_wind_speed= []
    for month in range(1,13):
        
        if month<10:
            browser.get("https://en.tutiempo.net/climate/0{}-{}/ws-421820.html".format(month, year))

            html = browser.page_source

            soup = BeautifulSoup(html, "html.parser")

            table = soup.find_all("table")

            row = table[3].select("tr")
            



            if month == 2:
                j = 29
                for i in row[1:j]:
                    Average_Temp.append(i.find_all("td")[1].text.strip())
                    Max_Temp.append(i.find_all("td")[2].text.strip())
                    Min_Temp.append(i.find_all("td")[3].text.strip())
                    Atm_Pressure.append(i.find_all("td")[4].text.strip())
                    Average_relative_humidity.append(i.find_all("td")[5].text.strip())
                    Total_rainfall_snowmelt.append(i.find_all("td")[6].text.strip())
                    Average_visibility.append(i.find_all("td")[7].text.strip())
                    Average_wind_speed.append(i.find_all("td")[8].text.strip())
                    Maximum_sustained_wind_speed.append(i.find_all("td")[9].text.strip())
                    
            elif month == 4 or month == 6 or month == 9:
                for i in row[1:31]:
                    Average_Temp.append(i.find_all("td")[1].text.strip())
                    Max_Temp.append(i.find_all("td")[2].text.strip())
                    Min_Temp.append(i.find_all("td")[3].text.strip())
                    Atm_Pressure.append(i.find_all("td")[4].text.strip())
                    Average_relative_humidity.append(i.find_all("td")[5].text.strip())
                    Total_rainfall_snowmelt.append(i.find_all("td")[6].text.strip())
                    Average_visibility.append(i.find_all("td")[7].text.strip())
                    Average_wind_speed.append(i.find_all("td")[8].text.strip())
                    Maximum_sustained_wind_speed.append(i.find_all("td")[9].text.strip())
                
                
            else:
                for i in row[1:32]:
                    Average_Temp.append(i.find_all("td")[1].text.strip())
                    Max_Temp.append(i.find_all("td")[2].text.strip())
                    Min_Temp.append(i.find_all("td")[3].text.strip())
                    Atm_Pressure.append(i.find_all("td")[4].text.strip())
                    Average_relative_humidity.append(i.find_all("td")[5].text.strip())
                    Total_rainfall_snowmelt.append(i.find_all("td")[6].text.strip())
                    Average_visibility.append(i.find_all("td")[7].text.strip())
                    Average_wind_speed.append(i.find_all("td")[8].text.strip())
                    Maximum_sustained_wind_speed.append(i.find_all("td")[9].text.strip())
                

        else:
            browser.get("https://en.tutiempo.net/climate/{}-{}/ws-421820.html".format(month, year))

            html = browser.page_source

            soup = BeautifulSoup(html, "html.parser")

            table = soup.find_all("table")

            row = table[3].select("tr")
            if month == 11:
                 for i in row[1:31]:
                    Average_Temp.append(i.find_all("td")[1].text.strip())
                    Max_Temp.append(i.find_all("td")[2].text.strip())
                    Min_Temp.append(i.find_all("td")[3].text.strip())
                    Atm_Pressure.append(i.find_all("td")[4].text.strip())
                    Average_relative_humidity.append(i.find_all("td")[5].text.strip())
                    Total_rainfall_snowmelt.append(i.find_all("td")[6].text.strip())
                    Average_visibility.append(i.find_all("td")[7].text.strip())
                    Average_wind_speed.append(i.find_all("td")[8].text.strip())
                    Maximum_sustained_wind_speed.append(i.find_all("td")[9].text.strip())
                    
                    
                    
            else:
                for i in row[1:32]:
                    Average_Temp.append(i.find_all("td")[1].text.strip())
                    Max_Temp.append(i.find_all("td")[2].text.strip())
                    Min_Temp.append(i.find_all("td")[3].text.strip())
                    Atm_Pressure.append(i.find_all("td")[4].text.strip())
                    Average_relative_humidity.append(i.find_all("td")[5].text.strip())
                    Total_rainfall_snowmelt.append(i.find_all("td")[6].text.strip())
                    Average_visibility.append(i.find_all("td")[7].text.strip())
                    Average_wind_speed.append(i.find_all("td")[8].text.strip())
                    Maximum_sustained_wind_speed.append(i.find_all("td")[9].text.strip())




    data= pd.DataFrame({"Average_Temp":Average_Temp, "Max_Temp": Max_Temp, "Min_Temp" : Min_Temp,"Atm_Pressure":Atm_Pressure, "Average_relative_humidity(%)": Average_relative_humidity, "Total_rainfall_snowmelt(mm)":Total_rainfall_snowmelt, "Average_visibility": Average_visibility, "Average_wind_speed": Average_wind_speed, "Maximum_sustained_wind_speed": Maximum_sustained_wind_speed})
    master_data[year] = data
```

```python
bigdata = pd.concat([master_data[i] for i in master_data], axis = 0)
```
